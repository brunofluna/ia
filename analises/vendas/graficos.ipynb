```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [2,3,4,3]

plt.plot(x, y)
plt.show()
```

```python
from matplotlib import legend


plt.plot(x, y, label='dados')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.title('Grafico de linha')
plt.legend()
plt.show()
```

```python
#plt.plot(x, y, label='dados')
plt.plot(x, y, ls=':', lw='5', color='red')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.title('Grafico de linha pontilhada')
plt.yticks([1,2,4])
plt.xticks([1,2,4])
plt.legend()
plt.show()
```

```python
#plt.bar(x,y)
plt.bar(x, y, label='dados')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.title('Grafico de barras')
plt.legend()
plt.show()

plt.show()
```

```python
plt.scatter(x, y, label='dados', marker='v')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.title('Grafico de pontos de encontro')
plt.legend()
plt.show()
```

```python
figura = plt.figure(figsize=(20, 8))
figura.add_subplot(121)
plt.plot(x,y)
plt.title('Grafico de linha')

figura.add_subplot(122)
plt.bar(x,y)
plt.title('Grafico de Barras')

figura.add_subplot(223)
plt.scatter(x,y)
plt.title('Grafico de bolinhas')
```

```python
valores = [3,5,10,3,1,8,7,7,7,10,5,3]
plt.hist(valores, label='frequencia')
plt.xlabel('valore sugeridos')
plt.ylabel('qtd')
plt.legend()
plt.show()
```
