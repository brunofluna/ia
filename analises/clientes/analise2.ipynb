```python
from operator import index

# df = pd.concat([df_vendas1, df_vendas2], ignore_index=True) faz a mesma coisa que o append, mas é mais rápido e eficiente.
# a proxima linha de código faz a junção dos dois dataframes, df_vendas1 e df_vendas2, e ignora os índices originais, criando um novo índice sequencial.
# O parâmetro ignore_index=True é usado para criar um novo índice sequencial para o DataFrame resultante.
df = pd.concat([df_vendas1, df_vendas2], ignore_index=True)
df
```

```python
# trocar nome da coluna
df.rename(columns={'cidade':'uf'}, inplace=True)
df
```

```python
# excluindo a série nome
df.drop('nome', axis=1, inplace=True)

# convertendo datas para datetime
df['nascimento'] = pd.to_datetime(df['data_nasc'], errors='coerce')

# excluindo a coluna data_nasc
df.drop('data_nasc', axis=1, inplace=True)

# excluir nulos da coluna nascimento
df.dropna(subset=['nascimento'], inplace=True)
df.info()
```

```python
# aplicar a moda da uf nos valores nulos
df['uf'].fillna(df['uf'].mode()[0], inplace=True)
df.info()
```

```python
# aplicar a média da renda mensal nos valores nulos
df['renda_mensal'].fillna(df['renda_mensal'].mean(), inplace=True)
df.info()
```
